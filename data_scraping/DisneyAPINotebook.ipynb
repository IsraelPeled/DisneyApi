```python
import json
import requests
from skimage import io
import matplotlib.pyplot as plt
```

```python
url = "https://api.disneyapi.dev/character"
movielist = ["Pinocchio","Snow White and the Seven Dwarfs","Tarzan","The Emperor's New Groove", "The lion King"]
```

### Functions

```python
def loadMoviefilteredData(film):
    res = requests.get(url+f'?films={film}')
    data = res.json()
    return data['data']
```

```python
def makeLists(data):
    ids = []
    names = []
    imageUrls = []
    films = []
    for index in data:
        ids.append(index['_id'])
        names.append(index['name'])
        imageUrls.append(index['imageUrl'])
        films.append(index['films'][0])
    return ids, films, names, imageUrls
```

```python
def makeJsonList(ids: list, films: list, names: list, imageUrls: list, jsonList: list, length: int):
    for i, (film, name, imageUrl) in enumerate(zip(films, names, imageUrls), start=length):
        obj = {'id': i,'film': film, 'name': name, 'imageUrl': imageUrl, 'score': 1400}
        jsonList.append(obj)
    return jsonList
```

```python
def makeJsonFile(jsonList: list):
    with open(f'characters.json', 'w') as json_file:
        json.dump(jsonList, json_file)
```

```python
def makeJsons(movielist: list):
    jsonList = []
    length = len(jsonList)
    for movie in movielist:
        data = loadMoviefilteredData(movie)
        ids, films, names, imageUrls = makeLists(data)
        jsonList = makeJsonList(ids, films, names, imageUrls, jsonList, len(jsonList))
    makeJsonFile(jsonList)
    return jsonList
```

## Run

```python
jsonList = makeJsons(movielist)
jsonList
```

```python
def reset_ids(data):
    for i, item in enumerate(data):
        item["id"] = i
    return data
```

```python
def load_json_array(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    return data
```
